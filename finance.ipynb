```python
import pandas as pd

df = pd.read_csv(r"C:\Users\Chinmoyee Bhuyan\OneDrive\Desktop\Finance Project\finance.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.drop("Unnamed: 0", axis=1)
```

```python
df.head()
```

```python
df["default"] = df["default"].map({"Yes": 1, "No": 0})
```

```python
df["student"] = df["student"].map({"Yes": 1, "No": 0})
df.head()
```

```python
df["default"].value_counts(normalize=True)
```

```python
df.groupby("default")["income"].mean()
```

```python
df.groupby("default")["balance"].mean()
```

```python
df.groupby("student")["default"].mean()
```

```python
df.to_csv("cleaned_finance.csv", index=False)
```

```python
import sqlite3
```

```python
conn = sqlite3.connect("finance.db")
```

```python
df.to_sql("finance", conn, if_exists="replace", index=False)
```

```python
#default count
pd.read_sql("""
SELECT "default", COUNT(*) 
FROM finance 
GROUP BY "default"
""", conn)
```

```python
pd.read_sql("""
SELECT "default", AVG(balance) 
FROM finance 
GROUP BY "default"
""", conn)
```

```python
pd.read_sql("""
SELECT student, AVG("default") 
FROM finance 
GROUP BY student
""", conn)
```
